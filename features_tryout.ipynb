```python
#import libraries
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data
data=pandas.read_csv('./test_data/data_japon.csv',skipinitialspace=True)
data.columns=data.columns.str.strip()#remove all whitespaces from data headers
#formating date column to datetieme
data['date']=pandas.to_datetime(data['date'],unit='D',origin='1899/12/30')
```

```python
#graphing the data

# Set up the plotting style
sns.set(style="whitegrid")

# Example: Plotting a histogram for a column
plt.figure(figsize=(8,6))
sns.histplot(data['amount'], bins=20, kde=True)
plt.title('Distribution of Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Boxplot to visualize outliers and distribution
plt.figure(figsize=(8,6))
sns.boxplot(x=data['reason'], y=data['amount'])
plt.title('Boxplot of Amount by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```
